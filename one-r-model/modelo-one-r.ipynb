##### criação das regras do One-R

```python
# Função para criar regras e calcular a taxa de erro
def criar_regras_e_calcular_erro(tot_SIM, tot_NAO, nome_atributo, valores):
    regras = {}
    taxa_erro = 0

    for i, valor in enumerate(valores):
        if tot_SIM[i] >= tot_NAO[i]:
            regras[valor] = 1  # Classe mais frequente é 1 (SIM)
            taxa_erro += tot_NAO[i]  # Erro é o número de NÃO
        else:
            regras[valor] = 0  # Classe mais frequente é 0 (NÃO)
            taxa_erro += tot_SIM[i]  # Erro é o número de SIM

    return regras, taxa_erro

# Valores únicos de cada atributo
valores_idade = ["20-40", "40-60", "60-80"]
valores_pressao = ["normal", "elevada", "alta"]
valores_colesterol = ["normal", "limítrofe", "alto"]
valores_fc_maxima = ["baixa", "normal", "alta"]
valores_depressao_st = ["nenhuma", "leve", "moderada"]

# Criar regras e calcular taxas de erro para cada atributo
regras_idade, taxa_erro_idade = criar_regras_e_calcular_erro(tot_idade_SIM, tot_idade_NAO, "idade", valores_idade)
regras_pressao, taxa_erro_pressao = criar_regras_e_calcular_erro(tot_pressao_SIM, tot_pressao_NAO, "pressao_repouso", valores_pressao)
regras_colesterol, taxa_erro_colesterol = criar_regras_e_calcular_erro(tot_colesterol_SIM, tot_colesterol_NAO, "colesterol", valores_colesterol)
regras_fc_maxima, taxa_erro_fc_maxima = criar_regras_e_calcular_erro(tot_fc_maxima_SIM, tot_fc_maxima_NAO, "fc_maxima", valores_fc_maxima)
regras_depressao_st, taxa_erro_depressao_st = criar_regras_e_calcular_erro(tot_depressao_st_SIM, tot_depressao_st_NAO, "depressao_st", valores_depressao_st)

# Exibir as regras e taxas de erro
print("Regras para Idade:", regras_idade)
print("Taxa de erro para Idade:", taxa_erro_idade)
print()

print("Regras para Pressão em Repouso:", regras_pressao)
print("Taxa de erro para Pressão em Repouso:", taxa_erro_pressao)
print()

print("Regras para Colesterol:", regras_colesterol)
print("Taxa de erro para Colesterol:", taxa_erro_colesterol)
print()

print("Regras para Frequência Cardíaca Máxima:", regras_fc_maxima)
print("Taxa de erro para Frequência Cardíaca Máxima:", taxa_erro_fc_maxima)
print()

print("Regras para Depressão ST:", regras_depressao_st)
print("Taxa de erro para Depressão ST:", taxa_erro_depressao_st)
print()

# Escolher o atributo com a menor taxa de erro
taxas_erro = {
    "idade": taxa_erro_idade,
    "pressao_repouso": taxa_erro_pressao,
    "colesterol": taxa_erro_colesterol,
    "fc_maxima": taxa_erro_fc_maxima,
    "depressao_st": taxa_erro_depressao_st
}

melhor_atributo = min(taxas_erro, key=taxas_erro.get)
menor_taxa_erro = taxas_erro[melhor_atributo]

print("Melhor atributo:", melhor_atributo)
print("Menor taxa de erro:", menor_taxa_erro)
```

##### utilizando a melhor regra para fazer as previsões

```python
# Regra para depressao_st
regra_depressao_st = {
    'nenhuma': 0,
    'leve': 0,
    'moderada': 1
}


# Aplicar a regra para fazer as previsões
df['previsao'] = df['depressao_st'].map(regra_depressao_st)

# Verificar as previsões
print(df[['depressao_st', 'previsao', 'target']].head())
```

```python
# Calcular a taxa de erro
taxa_erro = (df['previsao'] != df['target']).sum()
print(f"Taxa de erro: {taxa_erro}")
```

```python
from sklearn.metrics import accuracy_score
# Calcular a acurácia
acuracia = accuracy_score(df['target'], df['previsao'])
print("Acurácia do modelo: {:.2f}%".format(acuracia * 100))
```
